```python
lst = list()

with open('Kargil_War.htm', 'r') as reader:
    lst.append(reader.read())

text = ' '.join(map(str, lst))
```

```python
import re
cleanText = re.sub('[^A-Za-z0-9 ]+','', text)
```

```python
cleanText
```

```python
tokens = nltk.word_tokenize(cleanText)
```

```python
tokens
```

```python
tagged = nltk.pos_tag(tokens=tokens)
```

```python
tagged
```

```python
fdist = nltk.FreqDist(tokens)

filtered_word_freq = dict((word, freq) for word, freq in fdist.items() if not word.isdigit())

print(filtered_word_freq)
```

```python
from nltk.corpus import stopwords
  
stop_words = set(stopwords.words('english'))
  
filtered_sentence = [w for w in tokens if not w.lower() in stop_words]
  
filtered_sentence = list()
  
for w in tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
  
print(tokens)
print(filtered_sentence)
```

```python
unique_words = list()
unique_words_without_stop_words = list()

for w in tokens:
    if w not in unique_words:
        unique_words.append(w)

for w in filtered_sentence:
    if w not in unique_words_without_stop_words:
        unique_words_without_stop_words.append(w)
```

```python
print("Length of document:", len(tokens))
```

```python
unique_words
```

```python
unique_words_without_stop_words
```

```python
len(tagged), len(unique_words), len(unique_words_without_stop_words)
```

```python
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()

lemmatized_words = list()

for i in unique_words_without_stop_words:
    lemmatized_words.append((i, lemmatizer.lemmatize(i)))
```

```python
lemmatized_words
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer  = PorterStemmer()

stemmed_words = list()

for i in unique_words_without_stop_words:
    stemmed_words.append((i, porter_stemmer.stem(i)))
```

```python
stemmed_words
```

```python
pos_tagging = list()
tags = list()

for i in tagged:
    pos_tagging.append(i[0])
    tags.append(i[1])
```

```python
from tabulate import tabulate

final_lst = list()
op_names = ['Unique Words Without Stop Words', 'tokenization', 'lemmatization', 'pos_tagging', 'tag']

print("Length of document:", len(tokens))
print('Length of unique words:', len(unique_words))
print('Length of unique words without stop words:', len(unique_words_without_stop_words))

for i in range(len(unique_words_without_stop_words)):
    lst = list()
    lst.append(unique_words_without_stop_words[i])
    lst.append(tokens[i])
    lst.append(lemmatized_words[i])
    lst.append(stemmed_words[i])
    lst.append(pos_tagging[i])
    lst.append(tags[i])
    final_lst.append(lst)

print(tabulate(final_lst, headers = op_names))
```
