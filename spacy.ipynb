```python
nlp = spacy.load('en_core_web_sm')
```

```python
lst = list()

with open('Kargil_War.htm', 'r') as reader:
    lst.append(reader.read())
```

```python
text = ' '.join(map(str, lst))
```

```python
len(lst)
```

```python
len(text)
```

```python
text = text[200:20000]
```

```python
doc = nlp(text)
```

```python
print('Noun phrases:', [chunk.text for chunk in doc.noun_chunks])
print('Verbs:', [token.lemma_ for token in doc if token.pos_ == 'VERB'])
```

```python
for entity in doc.ents:
    print(entity.text, entity.label_)
```

```python
tokenization = list()
lemmatization = list()
pos_tagging = list()
tag = list()
dependencies = list()
shape = list()
is_alpha = list()
is_stop = list()

for token in doc:
    tokenization.append(token.text)
    lemmatization.append(token.lemma_)
    pos_tagging.append(token.pos_)
    tag.append(token.tag_)
    dependencies.append(token.dep_)
    shape.append(token.shape_)
    is_alpha.append(token.is_alpha)
    is_stop.append(token.is_stop)
```

```python
len(tokenization), len(lemmatization), len(pos_tagging), len(tag), len(dependencies), len(shape), len(is_alpha), len(is_stop)
```

```python
from tabulate import tabulate

final_lst = list()
op_names = ['tokenization', 'lemmatization', 'pos_tagging', 'tag', 'dependencies', 'shape', 'is_alpha', 'is_stop']

for i in range(3857):
    lst = list()
    lst.append(tokenization[i])
    lst.append(lemmatization[i])
    lst.append(pos_tagging[i])
    lst.append(tag[i])
    lst.append(dependencies[i])
    lst.append(shape[i])
    lst.append(is_alpha[i])
    lst.append(is_stop[i])
    final_lst.append(lst)

print(tabulate(final_lst, headers = op_names))
```
